```python
import pandas
```

```python
oceania_data = pandas.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
```

```python
oceania_data
```

```python
oceania_data.ix['Australia'].plot()
plt.xticks(rotation=45)
```

```python
oceania_data.plot()
```

```python
oceania_data.T.plot()
plt.ylabel('GDP per capita')
plt.xticks(rotation=45)
```

```python
plt.style.use('ggplot')
oceania_data.T.plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('GDP per capita')
```

```python
years = []
for col in oceania_data.columns:
    year = col[-4:]
    years.append(year)

gdp_australia = oceania_data.ix['Australia'].tolist()

print("years:", years)
print("gdp_australia:", gdp_australia)
plt.plot(years, gdp_australia, 'b-')
```

```python
years = []
for col in oceania_data.columns:
    year = col[-4:]
    years.append(year)

gdp_australia = oceania_data.ix['Australia'].tolist()
gdp_nz = oceania_data.ix['New Zealand'].tolist()

# print("years:", years)
# print("gdp_australia:", gdp_australia)
plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label='New Zealand')

plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita')
```

```python
plt.scatter(gdp_australia, gdp_nz)
plt.xlabel('Australia')
plt.ylabel('New Zealand')
```
